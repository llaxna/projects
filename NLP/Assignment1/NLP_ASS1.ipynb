```python
def clean_text(segment):
  text = re.sub("_", "", segment)
  text = re.sub('[^\w\s\.\?\-]',' ', text) # why space? because i noticed this: ( important—unimportant—unimportant—important— ) when — is removed all 4 become 1 word
  text = re.sub("\s+", " ", text)
  text = re.sub("\.{2,}","", text)
  text = re.sub("\?{2,}","", text)

  return text
```

```python
stop_words = stopwords.words('english')
stop_words
```

```python
words = word_tokenize(corpus)
filtered_text = []

for token in words:
  if token.lower() not in stop_words:
    filtered_text.append(token)

print(filtered_text)
print(words)
```

```python
corpus = re.sub(r'\s+([\.!?\-”’])', r'\1', corpus) #to remove extra spaces before punctations
corpus
```

```python
len(corpus)
```

```python
segments = sent_tokenize(corpus.lower())
segments
```

```python
len(segments)
```

```python
wordFreq = {}
words = re.findall(r'\b\w+\b|[\.\?\-]', corpus.lower())
for word in words:
    wordFreq[word] = wordFreq.get(word, 0) + 1

print("Word Frequencies:")
for word, freq in wordFreq.items():
    print(f"{word}: {freq}")
```

```python
freq = pd.DataFrame(list(wordFreq.items()), columns=['Word', 'Frequency'])
freq
```

```python
bigram = []
for segment in segments:
  words = ['<s>'] + word_tokenize(segment) + ['</s>']
  bigram.append(list(ngrams(words,2)))
```

```python
bigram
```

```python
biModel = {}

for bi in bigram:
  for b in bi:
    context = (b[0])
    next = b[1]
    if context not in biModel:
      biModel[context] = {}

    if next not in biModel[context]:
        biModel[context][next] = 1
    else:
        biModel[context][next] += 1
```

```python
biModel
```

```python
data = [{'Input': context, 'output': next, 'freq': freq}
        for context, nextFreq in biModel.items()
        for next, freq in nextFreq.items()]

freqbi = pd.DataFrame(data)
freqbi
```

```python
trigram = []
for segment in segments:
  words = ['<s>'] + word_tokenize(segment) + ['</s>']
  trigram.append(list(trigrams(words)))
```

```python
trigram
```

```python
triModel = {}

for tri in trigram:
  for t in tri:
    context = (t[0], t[1])
    next = t[2]
    if context not in triModel:
      triModel[context] = {}

    if next not in triModel[context]:
        triModel[context][next] = 1
    else:
        triModel[context][next] += 1
```

```python
triModel
```

```python
data = [{'Input': context, 'output': next, 'freq': freq}
        for context, nextFreq in triModel.items()
        for next, freq in nextFreq.items()]

freqTri = pd.DataFrame(data)
freqTri
```

```python
laplace = {}
vocabSize = len(freq['Word'])

for index, row in freqTri.iterrows():
    trigram = row['Input']
    numerator = row['freq'] + 1

    biCount = freqbi[(freqbi['Input'] == trigram[0]) & (freqbi['output'] == trigram[1])]['freq'].values[0]

    denominator = biCount + vocabSize
    probability = numerator / denominator

    if trigram not in laplace:
        laplace[trigram] = {}
    laplace[trigram][row['output']] = probability

laplace
```

```python
data = [{'Input': context, 'output': next, 'prob (laplace)': prob}
        for context, next_word_freq in laplace.items()
        for next, prob in next_word_freq.items()]
freqLap = pd.DataFrame(data)
freqLap
```

```python
genSent = []
for i in range(5):
  print("\n\tGenerating sentence",i+1,"\n:")
  first = input("Enter first word: ").lower()
  second = input("Enter second word: ").lower()
  sentence = [first,second]
  while len(sentence) <= 10:
      context = tuple(sentence[-2:])

      if context not in laplace:
        break

      next = max(laplace[context], key=laplace[context].get)
      sentence.append(next)

  generatedText = ' '.join(sentence)
  genSent.append(generatedText)
  print("generated text:", generatedText)
```

```python
def perplexityCalc(test_set, triModel):
    N = len(test_set)
    logSum = 0

    for i in range(2, N):
        trigram = (test_set[i-2], test_set[i-1], test_set[i])
        if trigram in triModel:
            triProb = triModel[trigram[0:2]][trigram[2]]
        else:
            triProb = 1 / len(triModel)

        logSum += math.log(triProb)

    perplexity = math.exp(-logSum / N)
    return perplexity

print("Perplexity:", perplexityCalc(genSent, laplace))
```
