### Define the sigmoid function for logistic regression.
\begin{equation*}
sig(t)= \frac{1} {1 + e^{-t}}
\end{equation*}
where
$t=\mathbf{x}^\intercal \mathbf{\beta}$



```python
def sigmoid(t):
    return 1 / (1 + np.exp(-t))
```

### Define the Cost function for Logistic Regression

### $J(\beta ) = \frac{1}{m} \sum_{i=1}^m[-y_i log(sig_\theta (t_i) – (1 – y_i) log(1-sig_\theta (t_i))]$

```python
def computeCost(beta, X, y):
## Write the code that calculates the cross entroy loss (cost function of logisitc regression)

    t = X.dot(beta)
    m = len(y)

    J = 1/m * np.sum(-y*np.log(sigmoid(t))-(1-y)*np.log(1-sigmoid(t)))
    return J
```

### Write the gradient descent function
### $\beta_j = \beta_j -\alpha \frac{1}{m} \sum_{i=1}^m((sig(\mathbf{x}^\intercal \mathbf{\beta}) - y).x_j)$

```python
len(y_train)
```

```python
def gradientDescent(x,y,beta,alpha,niter):
## Write the code for the gradient descent algorithm that takes X and the multiclass labels
## and returns the beta matrix and cost history

    cost_history = []

    for i in range(niter):
      beta_temp = beta.copy()
      y_pred = sigmoid(x.dot(beta))

      for j in range(beta.shape[1]):
          gradient = (1 / len(y)) * np.sum((y_pred[:, j] - y[:, j]).reshape(-1, 1) * x, axis=0)
          beta[:, j] = beta_temp[:,j] - alpha * gradient

      cost = computeCost(beta,x,y)
      cost_history.append(cost)

      print("iteration:",i)
      print("beta:",beta)
      print("cost:",cost)

    return beta, cost_history;
```
