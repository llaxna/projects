# Part 3:
## Gradient Descent Implementation
## First write the cost function implementation

### Cost Function

### $J(\beta ) = \frac {1} {m}\sum_{i=1}^m(f(x)-y_i)^2$
### $J(\beta ) = \frac {1} {m} (X \beta -y)^T (X \beta -y)$

```python
def computeCost(x,y,beta):
    #Compute the cost of a particular choice of beta.
    #use the first or second formula
    a = (x.dot(beta)-y).T
    b = x.dot(beta) - y
    cost = a.dot(b)/y.size

    return cost;
```

### Second write the gradient descent function. Use the notes from lectrues as a reference for the code
### 1.5 points

### Repeat until convergence{
###         $\beta_j = \beta_j -\alpha \frac{1}{m} \sum_{i=1}^m((f(x) - y).x_j)$
###       for j =0,..n }
### update coeficients sumiltaneously

```python
print(P_matrix[:,1].min())
print(P_matrix[:,1].max())

print(P_matrix[:,2].min())
print(P_matrix[:,2].max())
```

```python
def gradientDescent(x,y,beta,alpha,niter): #3 betas randomly initialized first
    # Write your code here, refer to lecture notes for the algorithm
    # Return the final values of beta and an array of the cost for every iteration

    #x is the p_matrix
    #y is y_train (vector)
    #niter is num of iterations

    #used feature scaling (Standardization)

    x[:,1] = (x[:,1] - np.mean(x[:,1])) / np.std(x[:,1])
    x[:,2] = (x[:,2] - np.mean(x[:,2])) / np.std(x[:,2])

    cost_history = []

    for i in range(niter):

      y_pred = x.dot(beta)

      for j in range(beta.size):
        beta[j] = beta[j] - ((alpha * np.sum((y_pred - y) * x[:, j])) / y.size)

      cost = computeCost(x,y,beta)
      cost_history.append(cost)

      print("iteration:",i)
      print("beta:",beta)
      print("cost:",cost)

    return beta, cost_history; #history checks convergence
```
